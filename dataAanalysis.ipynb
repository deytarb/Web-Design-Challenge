```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Instructions/Resources/cities.csv")
data.head()
```

```python
temp = data[["Lat", "Max Temp"]]

plt.figure()
fig, ax = plt.subplots(1, 1)
temp.plot(
    kind='scatter',
    x='Lat',
    y='Max Temp',
    ax = ax,
    grid=True
)
fig.suptitle("City Lattitude vs. Max Temperatire")
plt.show()
```

```python
hum = data[["Lat", "Humidity"]]

plt.figure()
fig, ax = plt.subplots(1, 1)
hum.plot(
    kind='scatter',
    x='Lat',
    y='Humidity',
    ax = ax,
    grid=True
)
fig.suptitle("City Lattitude vs. Humidity")
```

```python
spd = data[["Lat", "Wind Speed"]]

plt.figure()
plt.figure()
fig, ax = plt.subplots(1, 1)
spd.plot(
    kind='scatter',
    x='Lat',
    y='Wind Speed',
    ax = ax,
    grid=True
)
fig.suptitle("City Lattitude vs. Wind Speed")
```

```python
clns = data[["Lat", "Cloudiness"]]

plt.figure()
plt.figure()
fig, ax = plt.subplots(1, 1)
clns.plot(
    kind='scatter',
    x='Lat',
    y='Cloudiness',
    ax = ax,
    grid=True
)
fig.suptitle("City Lattitude vs. Cloudiness")
```
